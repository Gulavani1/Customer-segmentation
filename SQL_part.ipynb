```python
#How many different types of stocks did the customer purchase?
query ="""
select online_transactions_cleaned.* 
from bootcamp.online_transactions_cleaned
"""
pd.read_sql(query, connect)
```

```python
#What is the most popular stock? (You can look at top 10 sold items - does this differ across markets?)

query ="""
select online_transactions_cleaned.stock_code,
sd.description, 
sum(quantity)as total_sold
from bootcamp.online_transactions_cleaned 
join bootcamp.stock_description sd on online_transactions_cleaned.stock_code = sd.stock_code
group by online_transactions_cleaned.stock_code,sd.description  
order by total_sold desc
limit 10"""
pd.read_sql(query, connect)
```

```python
#What is the average order value i.e. price * quantity?
query ="""
select customer_id,
country ,
round(avg(price*quantity),2) as average_order_value,
sum(quantity) as total_items_purchased,
count(distinct stock_code)as number_items_purchased
from bootcamp.online_transactions_cleaned
group by customer_id , country 
order by average_order_value desc"""
```

```python
pd.read_sql(query, connect)
```

```python
#How many Stocks have the Description “Unknown”? How will you handle this when building customer segments?


query = """
select 
online_transactions_cleaned.invoice,
       online_transactions_cleaned.stock_code,
       online_transactions_cleaned.customer_id,
       sd.description
from bootcamp.online_transactions_cleaned
left join (select *
           from bootcamp.stock_description
          ) sd on online_transactions_cleaned.stock_code = sd.stock_code
           where sd.description = 'Unknown' """

online_trans=pd.read_sql(query, connect)

```
